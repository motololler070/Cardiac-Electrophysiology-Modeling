## Overview

Heart pumping happesn due to contrations and releases of cardiomyocytes. On cellular level contraction are triggered via increase of level of calcium, which affects [tropomyosin complex](https://en.wikipedia.org/wiki/Sliding_filament_theory). [Raise of calcium](https://en.wikipedia.org/wiki/Calcium-induced_calcium_release#:~:text=Calcium%2Dinduced%20calcium%20release%20(CICR,endoplasmic%20reticulum%20or%20sarcoplasmic%20reticulum) is result of electrochemical reactions regulated via cellular ion channels and membrane potential. Simultaneous contraction is achieved through propogation of electrical signal through heart muscle fiber regulated via [Sinoatrial_node](https://en.wikipedia.org/wiki/Sinoatrial_node). Record of activation of cell membrane is called activation potential, which allows to study fitness of tissues and identify diseases. [Antiarrhythmic agents](https://en.wikipedia.org/wiki/Antiarrhythmic_agent) drugs helps to mitigate too intense and rapid activation, which are harmful for patients with heart diseases. Antiarrhythmic agents classified in 5 classes, based on target of impact:
- Class I agents interfere with the sodium (Na+) channel.
- Class II agents are anti-sympathetic nervous system agents. Most agents in this class are beta blockers.
- Class III agents affect potassium (K+) efflux.
- Class IV agents affect calcium channels and the AV node.
- Class V agents work by other or unknown mechanisms.

### Excitation-contraction stages

Electochemical mechanism of heart contraction is called [Cardiac excitation-contraction coupling](https://en.wikipedia.org/wiki/Cardiac_excitation-contraction_coupling), which happens in stages:
1. **Diastole**:
During this phase, the cell membrane of the myocardial cell is polarized with a negative charge relative to the outside. The resting membrane potential is maintained by the Na⁺/K⁺ ATPase pump, which transports Na⁺  ions out of the cell and K⁺ ions into the cell.
2. **Depolarization**:
When an action potential reaches the myocyte, **voltage-gated Na⁺ channels open rapidly**, allowing a large influx of sodium ions into the cell. This causes a rapid depolarization of the cell membrane, making the interior more positive.
3. **Plateau Phase**:
L-type calcium channels opens. This influx maintains the plateau phase of the action potential, keeping the cell depolarized.
4. **Calcium-Induced Calcium Release (CICR)**:
Entered calcium triggers the release of additional calcium from sarcoplasmic reticulum (SR). This amplifies the amount of available calcium in the cytoplasm.
5. **Contraction**:
Vast amount of relased calcium bind to troponin C. Myosin heads start to interact with actin filaments, forming cross-bridges and initiating the sliding filament mechanism of contraction.
6. **Repolarization**:
As action potential comes to end K⁺ channels open. This restores the negative internal charge of the cell, returning it to its resting state.
7. **Relaxation**:
Calcium is actively removed from the cytosol via the SR Ca²⁺ ATPase (SERCA) pump back into the SR and through the Na⁺/Ca²⁺ exchanger on the cell membrane. With reduced calcium levels, troponin no longer binds calcium, and the myosin heads detach from actin, leading to relaxation of the muscle fiber.

### Equations

In current work we study activation potential based on macroscopic transmembrane currents using [Hodkin-Huxley](https://en.wikipedia.org/wiki/Hodgkin%E2%80%93Huxley_model) model. Given equations connects conductance of channels $g$, capacitance of cell $C$ and potentials of membrane, potassium, calcium and soda $V$:
$$
    I = C_m \frac{d V_m}{dt} + g_K(V_m - V_K) + g_{Na}(V_m - V_{Na}) + g_l(V_m - V_l)
$$

More advanced approach yield through consideration of dependance of conductanse from time and voltage. New equation is given in form
$$
    I = C_m \frac{{\mathrm {d} }V_{m}}{{\mathrm {d} }t}+{\bar {g}}_{\text{K}}n^{4}(V_{m}-V_{K})+{\bar {g}}_{\text{Na}}m^{3}h(V_{m}-V_{Na})+{\bar {g}}_{l}(V_{m}-V_{l})
$$
, where $n,m,h$ are given by first order kinetics
$$
    \begin{align*}
        \frac{{dn}}{{dt}} &=\alpha_n(V_m)(1-n)-\beta_n(V_m)n,\\
        \frac{{dm}}{{dt}} &=\alpha_m(V_m)(1-m)-\beta_m(V_m)m,\\
        \frac{{dh}}{{dt}} &=\alpha_h(V_m)(1-h)-\beta_h(V_m)h.
    \end{align*}
$$

These equations are highly nonlinear and practially studied nearly experimentally achieved parametrs.


Before activation cell membrane voltage is kept through difference of ions concentration, which is given by [Nernst equation](https://en.wikipedia.org/wiki/Nernst_equation):

$$
    E_{K} = \frac{RT}{zF} \ln\left(\frac{[K^+]_o}{[K^+]_i}\right)
$$

Now let's run aforemention file with model [Beeler and Reuter 1977](https://pmc.ncbi.nlm.nih.gov/articles/PMC1283659/pdf/jphysiol00808-0190.pdf). This model  uses Hodkin-Huxley equations in form:
$$
     I = C_m \frac{dV_m}{dt} + I_{K_1} + I_{x_1} + I_{Na} + I_{Ca} - I_{external}
$$
,where
- $i_{Na}$ - an initial fast inward
current carried primarily by sodium, ;
-  $i_{CA}$ - a secondary or slow inward
current, carried mainly but not exclusively by calcium ions;
- $I_{x1}$ - a timeactivated outward current designated ;
- $I_{K_1}$ - time-independent potassium outward current designated;
- $I_{external}$ - constant background current.

Coefficients $\alpha$ and $\beta$ in equations for $n,m,t$ are given  in form
$$
    \alpha = \frac{0.012 * \exp(-0.008 * (V + 28))}{(\exp(0.15 * (V + 28)) + 1)}
$$


```python
import matplotlib.pyplot as plt
import myokit

m, p, x = myokit.load('br-1977.mmt')
s = myokit.Simulation(m, p)
d = s.run(1000)

plt.figure()
plt.plot(d.time(), d['membrane.V'])
plt.show()
```
